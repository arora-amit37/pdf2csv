```python
def normalize_ocr_text(text):
    """Apply common OCR error corrections to text."""
    if pd.isna(text):
        return ""
    text = str(text)
    
    # Common OCR character substitutions
    corrections = {
        'O': '0',
        'o': '0',
        'S': '5',
        's': '5',
        'I': '1',
        'l': '1',
        '|': '1',
        'B': '8',
        'b': '8',
        'Z': '2',
        'z': '2',
    }
    
    corrected = []
    for char in text:
        corrected.append(corrections.get(char, char))
    
    return ''.join(corrected)

def extract_age_from_raw_text(raw_text):
    """
    Extract Borr_Age from OCR text with robust error handling.
    Returns the first valid age found (61-95), or None if none found.
    """
    if pd.isna(raw_text):
        return None
    
    # Normalize OCR errors
    normalized = normalize_ocr_text(raw_text)
    
    # Extract all potential numeric tokens
    # Look for sequences of digits, possibly with separators/noise
    tokens = re.findall(r'\d+', normalized)
    
    valid_ages = []
    
    for token in tokens:
        try:
            value = int(token)
            # Check if in valid age range
            if 61 <= value <= 98:
                valid_ages.append(value)
        except (ValueError, OverflowError):
            continue
    
    # Return first valid age found (left-to-right ordering preserved)
    if valid_ages:
        return valid_ages[0]
    
    # If no valid ages found with basic extraction, try more aggressive cleaning
    # Look for 2-digit sequences that might be ages with noise
    potential_ages = re.findall(r'\b(\d{2})\b', normalized)
    
    for age_str in potential_ages:
        try:
            value = int(age_str)
            if 61 <= value <= 98:
                return value
        except (ValueError, OverflowError):
            continue
    
    # Try to find ages embedded in longer digit sequences
    # Extract all 2-digit substrings from longer numbers
    for token in tokens:
        if len(token) >= 2:
            for i in range(len(token) - 1):
                two_digit = token[i:i+2]
                try:
                    value = int(two_digit)
                    if 61 <= value <= 98:
                        return value
                except (ValueError, OverflowError):
                    continue
    
    return None

# Process rows where Borr_Age is NaN or out of valid range
for idx in df3.index:
    current_age = df3.at[idx, 'Borr_Age']
    
    # Check if row needs correction
    needs_correction = False
    
    if pd.isna(current_age):
        needs_correction = True
    else:
        try:
            age_val = float(current_age)
            if age_val < 61 or age_val > 98:
                needs_correction = True
        except (ValueError, TypeError):
            needs_correction = True
    
    # If row needs correction, attempt to extract from raw_text2
    if needs_correction:
        raw_text = df3.at[idx, 'raw_text']
        extracted_age = extract_age_from_raw_text(raw_text)
        
        if extracted_age is not None:
            df3.at[idx, 'Borr_Age'] = float(extracted_age)
```

## CoBorr_Age

```python
# 1. Basic counts
total_rows = len(df3)
num_nans = df3['Coborr_Age'].isna().sum()
pct_nans = (num_nans / total_rows) * 100

# 2. Filter for non-NaN values to check ranges
valid_ages = df3['Coborr_Age'].dropna()

# 3. Inside Range (55 - 95)
inside_mask = (valid_ages >= 55) & (valid_ages <= 95)
num_inside = inside_mask.sum()
pct_inside = (num_inside / total_rows) * 100

# 4. Outside Range (< 55 or > 95)
num_outside = (~inside_mask).sum()
pct_outside = (num_outside / total_rows) * 100

# Display results
print(f"--- Statistics for Coborr_Age in df3 ---")
print(f"Total Rows: {total_rows}")
print(f"Number of NaNs: {num_nans} ({pct_nans:.2f}%)")
print(f"Values Inside 55-95: {num_inside} ({pct_inside:.2f}%)")
print(f"Values Outside 55-95: {num_outside} ({pct_outside:.2f}%)")
```

```python
def normalize_ocr_text(text):
    """Apply common OCR error corrections to text."""
    if pd.isna(text):
        return ""
    text = str(text)
    
    # Common OCR character substitutions
    corrections = {
        'O': '0',
        'o': '0',
        'S': '5',
        's': '5',
        'I': '1',
        'l': '1',
        '|': '1',
        'B': '8',
        'b': '8',
        'Z': '2',
        'z': '2',
    }
    
    corrected = []
    for char in text:
        corrected.append(corrections.get(char, char))
    
    return ''.join(corrected)

def extract_coborr_age_from_raw_text(raw_text, borr_cnt):
    """
    Extract Coborr_Age from OCR text with conservative logic.
    Only returns a value if Borr_Cnt indicates 2 borrowers AND two valid ages found.
    Returns None otherwise.
    """
    if pd.isna(raw_text):
        return None
    
    # Critical check: only proceed if Borr_Cnt is 2
    if pd.isna(borr_cnt):
        return None
    
    try:
        borr_cnt_val = int(borr_cnt)
        if borr_cnt_val != 2:
            # Single borrower - no co-borrower should exist
            return None
    except (ValueError, TypeError):
        return None
    
    # Normalize OCR errors
    normalized = normalize_ocr_text(raw_text)
    
    # Extract all potential numeric tokens
    tokens = re.findall(r'\d+', normalized)
    
    valid_ages = []
    
    for token in tokens:
        try:
            value = int(token)
            # Check if in valid age range
            if 61 <= value <= 95:
                valid_ages.append(value)
        except (ValueError, OverflowError):
            continue
    
    # Must find exactly 2 or more valid ages for co-borrower to exist
    if len(valid_ages) >= 2:
        # Return second age (first is borrower, second is co-borrower)
        return valid_ages[1]
    
    # If no valid ages found with basic extraction, try 2-digit boundaries
    potential_ages = re.findall(r'\b(\d{2})\b', normalized)
    
    valid_ages_alt = []
    for age_str in potential_ages:
        try:
            value = int(age_str)
            if 61 <= value <= 95:
                valid_ages_alt.append(value)
        except (ValueError, OverflowError):
            continue
    
    if len(valid_ages_alt) >= 2:
        return valid_ages_alt[1]
    
    # Try embedded 2-digit sequences as last resort
    embedded_ages = []
    for token in tokens:
        if len(token) >= 2:
            for i in range(len(token) - 1):
                two_digit = token[i:i+2]
                try:
                    value = int(two_digit)
                    if 61 <= value <= 95:
                        embedded_ages.append(value)
                        if len(embedded_ages) >= 2:
                            return embedded_ages[1]
                except (ValueError, OverflowError):
                    continue
    
    # Not enough valid ages found - no co-borrower age available
    return None

# Process rows where Coborr_Age is NaN or out of valid range
for idx in df3.index:
    current_coborr_age = df3.at[idx, 'Coborr_Age']
    
    # Check if row needs correction
    needs_correction = False
    
    if pd.isna(current_coborr_age):
        needs_correction = True
    else:
        try:
            coborr_age_val = float(current_coborr_age)
            if coborr_age_val < 61 or coborr_age_val > 95:
                needs_correction = True
        except (ValueError, TypeError):
            needs_correction = True
    
    # If row needs correction, attempt to extract from raw_text2
    if needs_correction:
        raw_text = df3.at[idx, 'raw_text']
        borr_cnt = df3.at[idx, 'Borr_Cnt']
        
        coborr_age_extracted = extract_coborr_age_from_raw_text(raw_text, borr_cnt)
        
        # Only populate if we have a valid extraction
        if coborr_age_extracted is not None:
            df3.at[idx, 'Coborr_Age'] = float(coborr_age_extracted)
        else:
            # Explicitly set to NaN if Borr_Cnt is 1 or no valid second age found
            df3.at[idx, 'Coborr_Age'] = pd.NA
            
# This removes the .0 and keeps NaNs as <NA>
df3['Coborr_Age'] = pd.to_numeric(df3['Coborr_Age'], errors='coerce').astype('Int64')
```

## hecm_margin

```python
# 1. Basic counts
total_rows = len(df3)
num_nans = df3['hecm_margin'].isna().sum()
pct_nans = (num_nans / total_rows) * 100

# 2. Filter for non-NaN values to check ranges
valid_ages = df3['hecm_margin'].dropna()

# 3. Inside Range (55 - 95)
inside_mask = (valid_ages >= 0) & (valid_ages <= 3.25)
num_inside = inside_mask.sum()
pct_inside = (num_inside / total_rows) * 100

# 4. Outside Range (< 55 or > 95)
num_outside = (~inside_mask).sum()
pct_outside = (num_outside / total_rows) * 100

# Display results
print(f"--- Statistics for hecm_margin in df3 ---")
print(f"Total Rows: {total_rows}")
print(f"Number of NaNs: {num_nans} ({pct_nans:.2f}%)")
print(f"Values Inside 55-95: {num_inside} ({pct_inside:.2f}%)")
print(f"Values Outside 55-95: {num_outside} ({pct_outside:.2f}%)")
```

```python
def normalize_ocr_for_hecm_margin(text):
    """
    Aggressively normalize OCR text to recover numeric values.
    Handles common OCR misreads and formatting issues.
    """
    if pd.isna(text):
        return ""
    
    text = str(text)
    
    # Replace common OCR letter-to-digit mistakes
    ocr_corrections = {
        'O': '0',
        'o': '0',
        'Q': '0',
        'D': '0',
    }
    
    corrected = []
    for char in text:
        corrected.append(ocr_corrections.get(char, char))
    
    text = ''.join(corrected)
    
    # Replace comma with dot for decimal separator
    text = text.replace(',', '.')
    
    return text

def extract_hecm_margin_from_raw_text(raw_text, pd_stmln_flg_value):
    """
    Extract hecm_margin from raw_text2 using structural assumptions.
    
    The pattern is: [int_rt] [int_rt_10yr] [hecm_margin] [pd_stmln_flg]
    
    Strategy:
    1. Locate pd_stmln_flg in the text (single capital letter: N, C, R, or H)
    2. Search backwards from that position to find the nearest valid decimal
    3. Validate range (0.0 - 3.25)
    """
    if pd.isna(raw_text):
        return None
    
    # Normalize OCR errors
    normalized = normalize_ocr_for_hecm_margin(raw_text)
    
    # Determine the pd_stmln_flg character to search for
    flag_char = None
    if not pd.isna(pd_stmln_flg_value):
        flag_str = str(pd_stmln_flg_value).strip().upper()
        if flag_str in ['N', 'C', 'R', 'H']:
            flag_char = flag_str
    
    # If we don't know the flag, try to find any of the valid flags
    valid_flags = ['N', 'C', 'R', 'H']
    
    # Find all positions of valid flag characters
    flag_positions = []
    for i, char in enumerate(normalized):
        if char in valid_flags:
            flag_positions.append(i)
    
    # If we have a specific flag, filter to only those positions
    if flag_char:
        flag_positions = [i for i in flag_positions if normalized[i] == flag_char]
    
    # Try each flag position (rightmost first, as it's most likely to be correct)
    for flag_pos in reversed(flag_positions):
        # Extract substring before the flag
        text_before_flag = normalized[:flag_pos]
        
        # Find all decimal numbers before the flag
        # Pattern: optional digits, dot, digits OR just digits
        decimal_pattern = r'(\d+\.?\d*|\d*\.\d+)'
        matches = list(re.finditer(decimal_pattern, text_before_flag))
        
        # Work backwards through matches to find the closest valid value
        for match in reversed(matches):
            candidate = match.group(1)
            
            # Clean up the candidate
            candidate = candidate.strip()
            
            # Skip empty or just-dot
            if not candidate or candidate == '.':
                continue
            
            try:
                value = float(candidate)
                
                # Check if in valid range for hecm_margin
                if 0.0 <= value <= 3.25:
                    # Round to 2 decimal places max
                    value = round(value, 2)
                    return value
            except (ValueError, OverflowError):
                continue
    
    # Fallback: try aggressive pattern matching without relying on flag position
    # Extract all decimal numbers from the entire text
    all_decimals = re.findall(r'\d+\.?\d*|\d*\.\d+', normalized)
    
    # Filter to valid range and take candidates
    valid_candidates = []
    for dec_str in all_decimals:
        if dec_str and dec_str != '.':
            try:
                val = float(dec_str)
                if 0.0 <= val <= 3.25:
                    valid_candidates.append(val)
            except (ValueError, OverflowError):
                continue
    
    # If we found valid candidates, prefer values that are common hecm_margin values
    # Common values are typically between 0.5 and 3.0
    if valid_candidates:
        # Sort by proximity to typical range (1.0 - 2.5)
        valid_candidates.sort(key=lambda x: abs(x - 1.75))
        return round(valid_candidates[0], 2)
    
    # Final fallback: try to extract numbers immediately before single capital letters
    pattern = r'(\d+\.?\d*|\d*\.\d+)\s*[\'"]?\s*[NCRH]'
    matches = re.findall(pattern, normalized)
    
    for match in reversed(matches):
        try:
            value = float(match)
            if 0.0 <= value <= 3.25:
                return round(value, 2)
        except (ValueError, OverflowError):
            continue
    
    return None

# Process rows where hecm_margin is invalid or missing
for idx in df3.index:
    current_hecm_margin = df3.at[idx, 'hecm_margin']
    
    # Determine if row needs correction
    needs_correction = False
    
    if pd.isna(current_hecm_margin):
        needs_correction = True
    else:
        try:
            margin_val = float(current_hecm_margin)
            if margin_val < 0.0 or margin_val > 3.25:
                needs_correction = True
        except (ValueError, TypeError):
            needs_correction = True
    
    # If correction needed, attempt extraction
    if needs_correction:
        raw_text = df3.at[idx, 'raw_text']
        
        # Try to get pd_strmln_flg, handle KeyError if column doesn't exist
        try:
            pd_stmln_flg = df3.at[idx, 'pd_strmln_flg']
        except KeyError:
            pd_stmln_flg = None
        
        extracted_margin = extract_hecm_margin_from_raw_text(raw_text, pd_stmln_flg)
        
        if extracted_margin is not None:
            # Format to max 2 decimal places
            if extracted_margin == int(extracted_margin):
                df3.at[idx, 'hecm_margin'] = int(extracted_margin)
            else:
                df3.at[idx, 'hecm_margin'] = extracted_margin
```

```python
def normalize_ocr_for_borr_cnt(text):
    """
    Normalize OCR text to handle common misreads for Borr_Cnt extraction.
    """
    if pd.isna(text):
        return ""
    
    text = str(text)
    
    # OCR character corrections for digits
    ocr_corrections = {
        'i': '1',
        'l': '1',
        'I': '1',
        '|': '1',
        '!': '1',
        'z': '2',
        'Z': '2',
    }
    
    corrected = []
    for char in text:
        corrected.append(ocr_corrections.get(char, char))
    
    return ''.join(corrected)

def extract_borr_cnt_from_raw_text(raw_text):
    """
    Extract Borr_Cnt from raw_text2 using structural assumptions.
    
    The pattern is: [Borr_Age] [Coborr_Age] [Borr_Cnt] [es_status]
    
    Strategy:
    1. Locate es_status keyword (Terminated, Assigned, or Endorsed)
    2. Search backwards to find the nearest single digit (1 or 2)
    3. That digit is Borr_Cnt
    """
    if pd.isna(raw_text):
        return None
    
    # Normalize OCR errors
    normalized = normalize_ocr_for_borr_cnt(raw_text)
    
    # Define es_status keywords to search for
    status_keywords = ['Terminated', 'Assigned', 'Endorsed', 'Termlnated', 'Termlnated', 
                       'Asslgned', 'Asslgned', 'Endarsed', 'Eodarsed']
    
    # Find all occurrences of status keywords (case insensitive)
    status_positions = []
    normalized_lower = normalized.lower()
    
    for keyword in status_keywords:
        keyword_lower = keyword.lower()
        pos = normalized_lower.find(keyword_lower)
        if pos != -1:
            status_positions.append(pos)
    
    # Also try partial matches for common OCR errors
    # Look for patterns like "Assigned", "Asslgned", etc.
    status_pattern = r'(Terminated|Assigned|Endorsed|Termlnated|Asslgned|Endarsed|Eodarsed)'
    status_matches = list(re.finditer(status_pattern, normalized, re.IGNORECASE))
    
    for match in status_matches:
        status_positions.append(match.start())
    
    # Sort positions and use the first (leftmost) occurrence
    if status_positions:
        status_positions.sort()
        first_status_pos = status_positions[0]
        
        # Extract text before the status keyword
        text_before_status = normalized[:first_status_pos]
        
        # Find all single digits (1 or 2) in the text before status
        # Looking for standalone digits, possibly with whitespace/punctuation around them
        digit_pattern = r'\b([12])\b|[^\d]([12])[^\d]|^([12])[^\d]|[^\d]([12])$'
        
        # Simpler approach: find all occurrences of '1' or '2' as separate tokens
        all_tokens = re.findall(r'\S+', text_before_status)
        
        # Work backwards through tokens to find the last occurrence of '1' or '2'
        for token in reversed(all_tokens):
            # Clean token of punctuation
            cleaned = re.sub(r'[^\d]', '', token)
            
            if cleaned in ['1', '2']:
                return int(cleaned)
        
        # Fallback: look for any '1' or '2' near the end of text_before_status
        # Search in the last 50 characters before status
        search_window = text_before_status[-50:] if len(text_before_status) > 50 else text_before_status
        
        # Find all '1' or '2' characters
        for i in range(len(search_window) - 1, -1, -1):
            if search_window[i] in ['1', '2']:
                return int(search_window[i])
    
    # If no status keyword found, try alternative approach
    # Look for pattern: [age] [age] [1 or 2] with ages being 2-digit numbers in 60-95 range
    age_pattern = r'\b([6-9]\d)\b'
    ages = re.findall(age_pattern, normalized)
    
    if len(ages) >= 1:
        # After finding ages, look for '1' or '2' that follows
        # Find position after last age
        last_age_match = None
        for match in re.finditer(age_pattern, normalized):
            last_age_match = match
        
        if last_age_match:
            text_after_ages = normalized[last_age_match.end():]
            
            # Look for first '1' or '2' in the next 20 characters
            search_text = text_after_ages[:20]
            
            for char in search_text:
                if char in ['1', '2']:
                    return int(char)
    
    return None

# Process rows where Borr_Cnt is NaN
for idx in df3.index:
    current_borr_cnt = df3.at[idx, 'Borr_Cnt']
    
    # Only process if Borr_Cnt is NaN
    if pd.isna(current_borr_cnt):
        raw_text = df3.at[idx, 'raw_text']
        
        # Attempt to extract from raw_text2
        extracted_cnt = extract_borr_cnt_from_raw_text(raw_text)
        
        if extracted_cnt is not None:
            df3.at[idx, 'Borr_Cnt'] = extracted_cnt

# Apply fallback rule: if Borr_Cnt still NaN, calculate from Coborr_Age
# Borr_Cnt = 2 if Coborr_Age is not NaN, otherwise 1
mask_still_nan = df3['Borr_Cnt'].isna()
df3.loc[mask_still_nan, 'Borr_Cnt'] = np.where(
    df3.loc[mask_still_nan, 'Coborr_Age'].notna(), 
    2, 
    1
)

# Convert Borr_Cnt to integer dtype
df3['Borr_Cnt'] = df3['Borr_Cnt'].astype('Int64')
```

```python
def normalize_ocr_for_zipcode(text):
    """
    Normalize OCR text to handle common misreads for zip code extraction.
    """
    if pd.isna(text):
        return ""
    
    text = str(text)
    
    # OCR character corrections for digits
    ocr_corrections = {
        'O': '0',
        'o': '0',
        'Q': '0',
        'q': '0',
        'I': '1',
        'i': '1',
        'l': '1',
        '|': '1',
        '!': '1',
        'S': '5',
        's': '5',
        'Z': '2',
        'z': '2',
        'B': '8',
        'b': '8',
    }
    
    corrected = []
    for char in text:
        corrected.append(ocr_corrections.get(char, char))
    
    return ''.join(corrected)

def extract_zipcode_from_raw_text(raw_text):
    """
    Extract prop_addr_zip_cd from raw_text2 using structural assumptions.
    
    The pattern is: ... [hecm_orgntn_fees] [prop_addr_zip_cd] [loan_typ]
    
    Strategy:
    1. Normalize OCR errors
    2. Find loan_typ pattern (two-digit strings like '02', '01', etc.)
    3. Search backwards to find a 5-digit number (the zip code)
    4. Validate that it falls in valid range (10000-99999)
    """
    if pd.isna(raw_text):
        return None
    
    # Normalize OCR errors
    normalized = normalize_ocr_for_zipcode(raw_text)
    
    # Define valid loan_typ values
    valid_loan_types = ['02', '01', '04', '05', '03', '06']
    
    # Find all occurrences of potential loan_typ patterns
    # Look for quoted or unquoted two-digit patterns
    loan_typ_pattern = r'["\']?0[0-6]["\']?'
    loan_typ_matches = list(re.finditer(loan_typ_pattern, normalized))
    
    # Also look for explicit valid loan types
    for loan_type in valid_loan_types:
        for match in re.finditer(re.escape(loan_type), normalized):
            loan_typ_matches.append(match)
    
    # Remove duplicates by position and sort by position
    unique_positions = {}
    for match in loan_typ_matches:
        pos = match.start()
        if pos not in unique_positions:
            unique_positions[pos] = match
    
    loan_typ_matches = sorted(unique_positions.values(), key=lambda m: m.start())
    
    # Try each loan_typ position (prefer rightmost as it's likely at end of row)
    for match in reversed(loan_typ_matches):
        loan_typ_pos = match.start()
        
        # Extract text before the loan_typ
        text_before_loan_typ = normalized[:loan_typ_pos]
        
        # Look for 5-digit sequences near the end of text_before_loan_typ
        # Search in the last 100 characters
        search_window = text_before_loan_typ[-100:] if len(text_before_loan_typ) > 100 else text_before_loan_typ
        
        # Find all 5-digit sequences
        five_digit_pattern = r'\b(\d{5})\b|(\d{5})'
        matches = list(re.finditer(five_digit_pattern, search_window))
        
        # Work backwards to find the closest 5-digit number to loan_typ
        for five_digit_match in reversed(matches):
            candidate = five_digit_match.group(0)
            
            try:
                zip_value = int(candidate)
                
                # Validate zip code range
                if 10000 <= zip_value <= 99999:
                    return zip_value
            except (ValueError, OverflowError):
                continue
    
    # Fallback: find all 5-digit sequences in entire text
    all_five_digits = re.findall(r'\d{5}', normalized)
    
    # Filter to valid zip code range
    for candidate in reversed(all_five_digits):
        try:
            zip_value = int(candidate)
            if 10000 <= zip_value <= 99999:
                return zip_value
        except (ValueError, OverflowError):
            continue
    
    # Additional fallback: look for 5-digit sequences with possible OCR noise
    # Extract all digit sequences and check if any are exactly 5 digits
    digit_sequences = re.findall(r'\d+', normalized)
    
    for seq in reversed(digit_sequences):
        if len(seq) == 5:
            try:
                zip_value = int(seq)
                if 10000 <= zip_value <= 99999:
                    return zip_value
            except (ValueError, OverflowError):
                continue
    
    return None

# Process rows where prop_addr_zip_cd is invalid or missing
for idx in df3.index:
    current_zip = df3.at[idx, 'prop_addr_zip_cd']
    
    # Determine if row needs correction
    needs_correction = False
    
    if pd.isna(current_zip):
        needs_correction = True
    else:
        try:
            zip_val = int(current_zip)
            if zip_val < 10000 or zip_val > 99999:
                needs_correction = True
        except (ValueError, TypeError):
            needs_correction = True
    
    # If correction needed, attempt extraction
    if needs_correction:
        raw_text = df3.at[idx, 'raw_text']
        
        extracted_zip = extract_zipcode_from_raw_text(raw_text)
        
        if extracted_zip is not None:
            df3.at[idx, 'prop_addr_zip_cd'] = extracted_zip

# Convert prop_addr_zip_cd to integer dtype where possible
# Use Int64 to allow NaN values if any remain
df3['prop_addr_zip_cd'] = df3['prop_addr_zip_cd'].astype('Int64')
```

```python
from datetime import datetime

def normalize_ocr_for_date(text):
    """
    Normalize OCR text to handle common misreads for date extraction.
    """
    if pd.isna(text):
        return ""
    
    text = str(text)
    
    # OCR character corrections for digits
    ocr_corrections = {
        'l': '1',
        'I': '1',
        '|': '1',
        'i': '1',
        'O': '0',
        'o': '0',
        'Q': '0',
        'q': '0',
        'S': '5',
        's': '5',
        'Z': '2',
        'z': '2',
        'B': '8',
        'b': '8',
    }
    
    corrected = []
    for char in text:
        corrected.append(ocr_corrections.get(char, char))
    
    return ''.join(corrected)

def is_valid_date(month, day, year):
    """
    Check if the given month, day, year form a valid date.
    """
    try:
        datetime(year, month, day)
        return True
    except (ValueError, OverflowError):
        return False

def extract_endrsmt_dt_from_raw_text(raw_text):
    """
    Extract endrsmt_dt from raw_text using structural assumptions.
    
    The endorsement date is typically:
    - After the lender name
    - The second date in the sequence (after clsing_dt, before tmntn_dt)
    - In M/D/YYYY format
    
    Strategy:
    1. Normalize OCR errors
    2. Find all date-like patterns (M/D/YYYY or MM/DD/YYYY)
    3. Validate each candidate
    4. Return the second valid date (if available)
    """
    if pd.isna(raw_text):
        return None
    
    # Normalize OCR errors
    normalized = normalize_ocr_for_date(raw_text)
    
    # Find all date-like patterns
    # Pattern: M/D/YYYY or MM/DD/YYYY with variations
    date_pattern = r'(\d{1,2})[/\-](\d{1,2})[/\-](\d{4})'
    date_matches = list(re.finditer(date_pattern, normalized))
    
    valid_dates = []
    
    for match in date_matches:
        month_str = match.group(1)
        day_str = match.group(2)
        year_str = match.group(3)
        
        try:
            month = int(month_str)
            day = int(day_str)
            year = int(year_str)
            
            # Validate date
            if is_valid_date(month, day, year):
                # Store as string in M/D/YYYY format
                date_str = f"{month}/{day}/{year}"
                valid_dates.append((date_str, match.start()))
        except (ValueError, OverflowError):
            continue
    
    # If we found at least 2 valid dates, return the second one (endrsmt_dt)
    if len(valid_dates) >= 2:
        return valid_dates[1][0]
    
    # If we only found 1 date, it might be the endorsement date
    # (clsing_dt might be missing or malformed)
    if len(valid_dates) == 1:
        return valid_dates[0][0]
    
    # Fallback: try more aggressive pattern matching
    # Look for dates with possible OCR noise around slashes
    relaxed_pattern = r'(\d{1,2})\s*[/\-\|]\s*(\d{1,2})\s*[/\-\|]\s*(\d{4})'
    relaxed_matches = list(re.finditer(relaxed_pattern, normalized))
    
    valid_dates_relaxed = []
    
    for match in relaxed_matches:
        month_str = match.group(1).strip()
        day_str = match.group(2).strip()
        year_str = match.group(3).strip()
        
        try:
            month = int(month_str)
            day = int(day_str)
            year = int(year_str)
            
            if is_valid_date(month, day, year):
                date_str = f"{month}/{day}/{year}"
                valid_dates_relaxed.append((date_str, match.start()))
        except (ValueError, OverflowError):
            continue
    
    if len(valid_dates_relaxed) >= 2:
        return valid_dates_relaxed[1][0]
    
    if len(valid_dates_relaxed) == 1:
        return valid_dates_relaxed[0][0]
    
    # Final fallback: extract all digit sequences and try to construct dates
    # Look for patterns like: number slash number slash 4-digit-year
    tokens = re.findall(r'\d+', normalized)
    
    # Try to find consecutive tokens that could form a date
    for i in range(len(tokens) - 2):
        try:
            month = int(tokens[i])
            day = int(tokens[i + 1])
            year = int(tokens[i + 2])
            
            # Check if year is 4 digits
            if len(tokens[i + 2]) == 4:
                if is_valid_date(month, day, year):
                    date_str = f"{month}/{day}/{year}"
                    # This is likely the first date, so continue looking
                    # Try next set
                    if i + 5 < len(tokens):
                        try:
                            month2 = int(tokens[i + 3])
                            day2 = int(tokens[i + 4])
                            year2 = int(tokens[i + 5])
                            
                            if len(tokens[i + 5]) == 4:
                                if is_valid_date(month2, day2, year2):
                                    return f"{month2}/{day2}/{year2}"
                        except (ValueError, OverflowError):
                            pass
                    
                    return date_str
        except (ValueError, OverflowError, IndexError):
            continue
    
    return None

# Process rows where endrsmt_dt is invalid or missing
for idx in df3.index:
    current_endrsmt_dt = df3.at[idx, 'endrsmt_dt']
    
    # Determine if row needs correction
    needs_correction = False
    
    if pd.isna(current_endrsmt_dt):
        needs_correction = True
    else:
        # Validate that it's a valid date string in M/D/YYYY format
        try:
            date_str = str(current_endrsmt_dt).strip()
            # Try to parse the date
            parts = date_str.split('/')
            if len(parts) == 3:
                month = int(parts[0])
                day = int(parts[1])
                year = int(parts[2])
                
                if not is_valid_date(month, day, year):
                    needs_correction = True
            else:
                needs_correction = True
        except (ValueError, AttributeError, IndexError):
            needs_correction = True
    
    # If correction needed, attempt extraction
    if needs_correction:
        raw_text = df3.at[idx, 'raw_text']
        
        extracted_date = extract_endrsmt_dt_from_raw_text(raw_text)
        
        if extracted_date is not None:
            df3.at[idx, 'endrsmt_dt'] = extracted_date
```

```python
def normalize_ocr_for_date(text):
    """
    Normalize OCR text to handle common misreads for date extraction.
    """
    if pd.isna(text):
        return ""
    
    text = str(text)
    
    # OCR character corrections for digits
    ocr_corrections = {
        'l': '1',
        'I': '1',
        '|': '1',
        'i': '1',
        'O': '0',
        'o': '0',
        'Q': '0',
        'q': '0',
        'S': '5',
        's': '5',
        'Z': '2',
        'z': '2',
        'B': '8',
        'b': '8',
    }
    
    corrected = []
    for char in text:
        corrected.append(ocr_corrections.get(char, char))
    
    return ''.join(corrected)

def is_valid_date(month, day, year):
    """
    Check if the given month, day, year form a valid date.
    """
    try:
        datetime(year, month, day)
        return True
    except (ValueError, OverflowError):
        return False

def extract_clsing_dt_from_raw_text(raw_text):
    """
    Extract clsing_dt from raw_text.
    
    The closing date is typically:
    - The first date appearing after the lender name
    - Before the endorsement date (second date)
    - In M/D/YYYY format
    
    Strategy:
    1. Normalize OCR errors
    2. Find all date-like patterns (M/D/YYYY or MM/DD/YYYY)
    3. Validate each candidate
    4. Return the first valid date
    """
    if pd.isna(raw_text):
        return None
    
    # Normalize OCR errors
    normalized = normalize_ocr_for_date(raw_text)
    
    # Find all date-like patterns
    # Pattern: M/D/YYYY or MM/DD/YYYY with variations
    date_pattern = r'(\d{1,2})[/\-](\d{1,2})[/\-](\d{4})'
    date_matches = list(re.finditer(date_pattern, normalized))
    
    valid_dates = []
    
    for match in date_matches:
        month_str = match.group(1)
        day_str = match.group(2)
        year_str = match.group(3)
        
        try:
            month = int(month_str)
            day = int(day_str)
            year = int(year_str)
            
            # Validate date
            if is_valid_date(month, day, year):
                # Store as string in M/D/YYYY format
                date_str = f"{month}/{day}/{year}"
                valid_dates.append((date_str, match.start()))
        except (ValueError, OverflowError):
            continue
    
    # Return the first valid date (clsing_dt)
    if len(valid_dates) >= 1:
        return valid_dates[0][0]
    
    # Fallback: try more aggressive pattern matching
    # Look for dates with possible OCR noise around slashes
    relaxed_pattern = r'(\d{1,2})\s*[/\-\|]\s*(\d{1,2})\s*[/\-\|]\s*(\d{4})'
    relaxed_matches = list(re.finditer(relaxed_pattern, normalized))
    
    valid_dates_relaxed = []
    
    for match in relaxed_matches:
        month_str = match.group(1).strip()
        day_str = match.group(2).strip()
        year_str = match.group(3).strip()
        
        try:
            month = int(month_str)
            day = int(day_str)
            year = int(year_str)
            
            if is_valid_date(month, day, year):
                date_str = f"{month}/{day}/{year}"
                valid_dates_relaxed.append((date_str, match.start()))
        except (ValueError, OverflowError):
            continue
    
    if len(valid_dates_relaxed) >= 1:
        return valid_dates_relaxed[0][0]
    
    # Additional fallback: look for dates without clear separators
    # Pattern: MMDDYYYY or M/D/YYYY with missing slashes
    no_separator_pattern = r'(\d{1,2})(\d{1,2})(\d{4})'
    
    # Try to find sequences that could be dates
    all_digit_sequences = re.findall(r'\d+', normalized)
    
    for seq in all_digit_sequences:
        # Try to parse as date with various interpretations
        if len(seq) == 8:  # MMDDYYYY
            try:
                month = int(seq[0:2])
                day = int(seq[2:4])
                year = int(seq[4:8])
                
                if is_valid_date(month, day, year):
                    return f"{month}/{day}/{year}"
            except (ValueError, OverflowError):
                pass
            
            # Try M/DD/YYYY
            try:
                month = int(seq[0:1])
                day = int(seq[1:3])
                year = int(seq[3:7])
                
                if is_valid_date(month, day, year):
                    return f"{month}/{day}/{year}"
            except (ValueError, OverflowError, IndexError):
                pass
        
        if len(seq) == 7:  # M/D/YYYY without separators
            try:
                month = int(seq[0:1])
                day = int(seq[1:2])
                year = int(seq[2:6])
                
                if is_valid_date(month, day, year):
                    return f"{month}/{day}/{year}"
            except (ValueError, OverflowError, IndexError):
                pass
    
    # Final fallback: extract all digit sequences and try to construct dates
    tokens = re.findall(r'\d+', normalized)
    
    # Try to find consecutive tokens that could form a date
    for i in range(len(tokens) - 2):
        try:
            month = int(tokens[i])
            day = int(tokens[i + 1])
            year = int(tokens[i + 2])
            
            # Check if year is 4 digits
            if len(tokens[i + 2]) == 4:
                if is_valid_date(month, day, year):
                    return f"{month}/{day}/{year}"
        except (ValueError, OverflowError, IndexError):
            continue
    
    return None

# Process all rows to extract clsing_dt from raw_text
for idx in df3.index:
    raw_text = df3.at[idx, 'raw_text']
    
    extracted_date = extract_clsing_dt_from_raw_text(raw_text)
    
    if extracted_date is not None:
        df3.at[idx, 'clsing_dt'] = extracted_date
```

```python
# ---------------------------
# OCR NORMALIZATION
# ---------------------------

def _normalize_ocr(text: str) -> str:
    if not isinstance(text, str):
        return ""

    t = text.upper()

    # Character-level OCR corrections
    char_map = {
        'O': '0',
        'Q': '0',
        'D': '0',
        'S': '5',
        'B': '8',
        'I': '1',
        'L': '1',
        ',': '.',
        '|': ' ',
        '’': ' ',
        "'": ' ',
        '"': ' ',
    }
    for k, v in char_map.items():
        t = t.replace(k, v)

    # Collapse whitespace
    t = re.sub(r'\s+', ' ', t)

    # Repair split decimals: "5 . 56", "5. 56", "5 .56"
    t = re.sub(r'(\d)\s*\.\s*(\d)', r'\1.\2', t)

    # Repair spaced digits: "5 56" -> "5.56"
    t = re.sub(r'\b([2-9])\s+([0-9]{2})\b', r'\1.\2', t)

    return t


# ---------------------------
# NUMERIC CANDIDATE GENERATION
# ---------------------------

def _generate_rate_variants(token: str):
    """
    Given a numeric-looking token, generate plausible float interpretations.
    """
    variants = set()

    try:
        f = float(token)
        variants.add(f)
    except Exception:
        pass

    # Missing decimal inference: "556" -> 5.56, "711" -> 7.11
    if token.isdigit() and len(token) == 3:
        variants.add(float(token[0] + '.' + token[1:]))

    # Two-digit integer like "18" -> 7.8 (OCR 78) or 6.8 etc is too risky, skip

    return variants


def _extract_numeric_tokens(text: str):
    """
    Extract numeric-like tokens with positions.
    """
    tokens = []
    for m in re.finditer(r'\b\d{1,4}(?:\.\d{1,3})?\b', text):
        tokens.append((m.group(), m.start()))
    return tokens


# ---------------------------
# POSITIONAL HELPERS
# ---------------------------

def _find_es_status_pos(text: str) -> int:
    m = re.search(r'\b(TERMINATED|ASSIGNED|ENDORSED)\b', text)
    return m.end() if m else 0


def _find_anchor_positions(text: str):
    """
    Approximate anchor positions for later fields to bound search.
    """
    anchors = {}
    for label, pat in {
        'pd_flag': r'\b[NCRH]\b',
        'rt_typ': r'\b[A MF]\b',
    }.items():
        m = re.search(pat, text)
        if m:
            anchors[label] = m.start()
    return anchors


# ---------------------------
# CORE SELECTION LOGIC
# ---------------------------

def _select_int_rate(norm_text, int_rt_10yr):
    es_pos = _find_es_status_pos(norm_text)
    anchors = _find_anchor_positions(norm_text)

    tokens = _extract_numeric_tokens(norm_text)

    candidates = []

    for tok, pos in tokens:
        for v in _generate_rate_variants(tok):
            if 2.0 <= v <= 9.5:
                candidates.append({
                    'val': v,
                    'pos': pos
                })

    if not candidates:
        return np.nan

    # Prefer values after es_status
    for c in candidates:
        c['after_es'] = c['pos'] >= es_pos

    # Prefer values before later anchors (pd_stmln_flg, rt_typ)
    for c in candidates:
        c['before_anchor'] = True
        for a in anchors.values():
            if c['pos'] > a:
                c['before_anchor'] = False
                break

    # Distance to int_rt_10yr if available
    for c in candidates:
        if pd.notna(int_rt_10yr):
            c['delta_10yr'] = abs(c['val'] - int_rt_10yr)
        else:
            c['delta_10yr'] = 0.0

    # Composite ranking
    candidates.sort(
        key=lambda c: (
            not c['after_es'],
            not c['before_anchor'],
            c['delta_10yr'],
            c['pos']
        )
    )

    return candidates[0]['val']


# ---------------------------
# ROW-LEVEL APPLY
# ---------------------------

def _fill_int_rt(row):
    cur = row['int_rt']
    if pd.notna(cur) and 2.0 <= cur <= 9.5:
        return cur

    norm = _normalize_ocr(row['raw_text'])
    return _select_int_rate(norm, row.get('int_rt_10yr', np.nan))


# ---------------------------
# APPLY UPDATE IN PLACE
# ---------------------------

mask = df3['int_rt'].isna() | ~df3['int_rt'].between(2.0, 9.5)
df3.loc[mask, 'int_rt'] = df3.loc[mask].apply(_fill_int_rt, axis=1)

```

```python
# ------------------------------------------------------------
# OCR NORMALIZATION
# ------------------------------------------------------------

def _normalize_ocr(text: str) -> str:
    if not isinstance(text, str):
        return ""

    t = text.upper()

    # Correct common OCR character misreads
    char_map = {
        'O': '0',
        'Q': '0',
        'I': '1',
        'L': '1',
        'S': '5',
        'Z': '2',
        '|': ' ',
        '/': ' ',
        '\\': ' ',
        '[': ' ',
        ']': ' ',
        '(': ' ',
        ')': ' ',
        "'": ' ',
        '"': ' ',
        ',': ' ',
    }
    for k, v in char_map.items():
        t = t.replace(k, v)

    # Normalize whitespace
    t = re.sub(r'\s+', ' ', t)

    # Remove stray non-numeric punctuation except decimal
    t = re.sub(r'[^0-9.\s]', ' ', t)
    t = re.sub(r'\s+', ' ', t)

    return t.strip()


# ------------------------------------------------------------
# NUMERIC TOKEN EXTRACTION
# ------------------------------------------------------------

def _extract_numeric_tokens(text: str):
    """
    Extract numeric-looking tokens with start positions.
    """
    tokens = []
    for m in re.finditer(r'\b\d{2,7}(?:\.\d+)?\b', text):
        tokens.append((m.group(), m.start()))
    return tokens


def _clean_numeric(token: str):
    """
    Convert token to float after fixing decimal artifacts.
    """
    try:
        return float(token)
    except ValueError:
        return None


# ------------------------------------------------------------
# CORE EXTRACTION LOGIC
# ------------------------------------------------------------

def _select_init_principal(norm_text: str):
    """
    Select init_prncpl_lmt using positional logic:
    AFTER max_clm_amt and BEFORE hecm_orgntn_fees.
    """
    tokens = _extract_numeric_tokens(norm_text)

    if not tokens:
        return np.nan

    # Convert tokens to numeric candidates
    candidates = []
    for tok, pos in tokens:
        val = _clean_numeric(tok)
        if val is None:
            continue
        candidates.append({'val': val, 'pos': pos})

    if not candidates:
        return np.nan

    # Filter plausible ranges (very permissive, accuracy later)
    candidates = [
        c for c in candidates
        if 10000 <= c['val'] <= 1000000
    ]
    if not candidates:
        return np.nan

    # Sort by position
    candidates.sort(key=lambda c: c['pos'])

    # Heuristic:
    # max_clm_amt usually appears first, init_prncpl_lmt next,
    # hecm_orgntn_fees is usually 0 or small integer afterward.
    #
    # Strategy:
    # - Identify large-ish number (max_clm_amt)
    # - Next large-ish number is init_prncpl_lmt
    # - Ignore trailing small values (fees, zip, loan type)

    large_vals = [c for c in candidates if c['val'] >= 30000]

    if len(large_vals) >= 2:
        # Prefer values divisible by 1000
        for i in range(len(large_vals) - 1):
            v = large_vals[i + 1]['val']
            if v % 1000 == 0:
                return float(v)

        # Fallback: second large number
        return float(large_vals[1]['val'])

    # Fallback: single plausible candidate
    if large_vals:
        return float(large_vals[0]['val'])

    return np.nan


# ------------------------------------------------------------
# ROW-LEVEL APPLY
# ------------------------------------------------------------

def _fill_init_prncpl_lmt(row):
    cur = row['init_prncpl_lmt']
    if pd.notna(cur):
        try:
            if float(cur) > 0:
                return cur
        except Exception:
            pass

    norm = _normalize_ocr(row['raw_text'])
    return _select_init_principal(norm)


# ------------------------------------------------------------
# APPLY UPDATE IN PLACE
# ------------------------------------------------------------

mask = df3['init_prncpl_lmt'].isna() | ~pd.to_numeric(
    df3['init_prncpl_lmt'], errors='coerce'
).notna()

df3.loc[mask, 'init_prncpl_lmt'] = (
    df3.loc[mask].apply(_fill_init_prncpl_lmt, axis=1)
)
```
